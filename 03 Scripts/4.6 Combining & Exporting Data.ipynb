### Concatenating Data Exercise

```python
# Create dictionaries of January 2020 data 

data1 = {'customer_id': ['6732', '767', '890', '635'],
        'month': ['Jan-20', 'Jan-20', 'Jan-20', 'Jan-20'],
        'purchased_meat' : [0, 13, 3, 4],
        'purchased_alcohol' : [1, 2, 10, 0],
        'purchased_snack': [10, 5, 1, 7]}
```

```python
# Create dictionaries of February 2020 data

data2 = {'customer_id': ['6732', '767', '890', '635'],
        'month': ['Feb-20', 'Feb-20', 'Feb-20', 'Feb-20'],
        'purchased_meat' : [0, 10, 5, 3],
        'purchased_alcohol' : [2, 4, 14, 0],
        'purchased_snack': [15, 3, 2, 6]}
```

```python
# Convert the data dictionaries to data frames

df = pd.DataFrame (data1, index = [0, 1, 2, 3])
df_1 = pd. DataFrame (data2, index = [0, 1, 2, 3])
```

```python
df
```

```python
df_1
```

```python
# Creating a list that contains the dataframes

frames = [df, df_1]
```

```python
frames
```

```python
# Concatenate the dataframes

df_concat = pd.concat (frames)
```

```python
# Checking the output

df_concat
```

```python
# Concatenate dataframes in wide-format

df_concat_w = pd.concat (frames, axis = 1)
```

```python
# Checking the outpot

df_concat_w
```

### Merging Data Exercise

```python
# Create a new dataframe with different columns

data3 = {'customer_id': ['6732', '767', '890', '635'],
        'month': ['Jan-20', 'Jan-20', 'Jan-20', 'Jan-20'],
        'days_purchased_on': [0, 10, 4, 1]}

df_2 = pd.DataFrame (data3, index = [0, 1, 2, 3])
```

```python
# Checking the new dataframe

df_2
```

```python
# Merge df and df_2 dataframe using 'customer_id' as key

df_merged = df.merge (df_2, on = 'customer_id')

df_merged
```

```python
# Merge df and df_2 dataframe using 'customer_id' and 'month' as key

df_merged_2 = df.merge (df_2, on = ['customer_id', 'month'])

df_merged_2
```

```python
# Merge df and df_2 dataframe using 'customer_id' and 'month' as key, and add a merge flag

df_merged_f = df.merge (df_2, on = ['customer_id', 'month'], indicator = True)

df_merged_f
```

```python
# Checking '_merge' value counts

df_merged_f['_merge'].value_counts()
```

```python
# Test merge without creating a new dataframe or overwriting

pd.merge (df, df_2, on = ['customer_id', 'month'], indicator = True)
```

```python
# Trying the how argument in merge

df.merge (df_2, on = ['customer_id', 'month'], how = 'inner')
```
