# 4.8 Grouping Data & Aggregating Variables

```python
# Import libraries

import pandas as pd
import numpy as np
import os
```

### 1. Import relevant data

```python
# Import orders_products_merged dataframe

path = r'/Users/queency_yustiawan/Study/CF/data analyst course/immersion/A4 exercise/Instacart Basket Analysis'

df_ords_prods_merge = pd.read_pickle (os.path.join (path, '02 Data', 'Prepared Data', 'orders_products_merge_derived.pkl'))
```

### 2. Find the aggregated mean of the 'order_number' column grouped by 'department_id'

```python
# Grouping data by 'department_id' and aggregate the 'order_number'

df_ords_prods_merge.groupby('department_id').agg({'order_number' : ['mean']})
```

### 3. Analyse the result

##### There are 21 department_id produced in the output, as opposed to only 8 department_id when we aggregate the subset.

### 4. Create the loyalty flag for existing customers

```python
# Create 'max_order' column

df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Checking output

df_ords_prods_merge.head(50)
```

```python
# Create loyalty flag with loc ( ) in 'loyalty_label' column

df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal Customer'
```

```python
df_ords_prods_merge.loc[ (df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular Customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New Customer'
```

```python
# Checking the frequency of the loyalty flag

df_ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Checking output

df_ords_prods_merge [['user_id', 'loyalty_flag', 'order_number', 'max_order']].head(30)
```

### 5. Check the basic statistic of prices based on the customers' loyalty

```python
# Checking min, max, mean of prices based on customers' loyalty

df_ords_prods_merge.groupby('loyalty_flag').agg({'prices':['min', 'max', 'mean']})
```

### 6. Create spending flag based on the customer's avg spending

```python
# Create 'spending_avg' column

df_ords_prods_merge['spending_avg'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Checking output

df_ords_prods_merge[['user_id', 'spending_avg']].head(20)
```

```python
# Create a flag column for the spending

df_ords_prods_merge.loc [df_ords_prods_merge ['spending_avg'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
df_ords_prods_merge.loc [df_ords_prods_merge ['spending_avg'] >= 10, 'spending_flag'] = 'High Spender'
```

```python
# Checking output

df_ords_prods_merge[['user_id', 'spending_avg', 'spending_flag']].head(30)
```

```python
# Count the frequency of the spending flag

df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

### 7. Create order frequency flag based on the customer's ordering behaviour

```python
# Count the median of each customer and create the "order frequency"

df_ords_prods_merge['order_frequency'] = df_ords_prods_merge.groupby(['user_id'])['order_interval_days'].transform(np.median)
```

```python
# Checking output

df_ords_prods_merge[['user_id', 'order_interval_days', 'order_frequency']].head(30)
```

```python
# Create a flag for the order frequency

df_ords_prods_merge.loc[df_ords_prods_merge['order_frequency'] > 20, 'frequency_flag'] = 'Non-frequenct customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['order_frequency'] > 10) & (df_ords_prods_merge['order_frequency'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['order_frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Checking frequency

df_ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
# Checking output

df_ords_prods_merge[['user_id', 'order_frequency', 'frequency_flag']].head(30)
```
